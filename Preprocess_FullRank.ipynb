# Preprocess

```python
import numpy as np
import pandas as pd
import LZH_Utilities as utl
```

### Sort Time Series Files

```python
# load checkpoint
T_running = utl.get_runtime_marker()

df_time_series = utl.read_time_series_data("all_dataset")

idx_size = df_time_series[0].shape[0]

# sort out the full rank data
idx_full_rank = []
for idx in np.arange(df_time_series[0].shape[0]):
    is_good = True
    
    for time in np.arange(9):
        if (not is_good):
            break
        for idx_col in np.arange(1, len(utl.DATAFRAME_COLUMN_TIME_SERIES)):
            if (np.isnan(df_time_series[time].iloc[idx][idx_col])):
                is_good = False
                break
    
    if (is_good):
        idx_full_rank.append(idx)
        
    if (idx % 2000 == 0):
        print("{0} out of {1} Finished, Runtime: {2}".format(
            idx, 
            idx_size, 
            utl.format_time_s_2_hms(utl.get_runtime_in_second(T_running))
        ))

print("total full rank sample: {0}".format(len(idx_full_rank)))
```

(Note: total full rank sample: 48750 (49061 without night_day dataset))

```python
df_time_series_full_rank = utl.create_empty_time_series_data()
for idx_time in np.arange(9):
    df_time_series_full_rank[idx_time] = df_time_series[idx_time].iloc[idx_full_rank]

# save checkpoint
utl.save_time_series_data("full_rank_dataset", df_time_series_full_rank)
```
